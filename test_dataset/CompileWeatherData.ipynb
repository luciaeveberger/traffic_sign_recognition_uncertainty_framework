```python
def create_compiled_file(file_path, header_line, data_type):
    """to compile files into a compiled_csv"""
    created_file = file_path + data_type 
    filenames = list()
    with open(d + created_file, 'w+') as outfile:
            outfile.write(header_line)
            for file in os.listdir(d + file_path):
                if file.endswith(".txt") and "produkt_" in file:
                    print(file)
                    with open(d + file_path+file) as infile:
                        for line in infile:
                            l = line.replace(' ', "")
                            if 'STATIONS_ID' not in l:
                                outfile.write(l)
            outfile.close()
```
